**Sequential Data**

* Sequential data is a type of data where the order of elements matters and **each element is dependent on or related to what comes before and/or after it**.
* Examples include:
  * Words in a sentence (where meaning depends on word order)
  * Time series data (where values follow a temporal sequence)
  * DNA sequences (where the order of nucleotides is crucial)
  * Music notes (where timing and order create melody)

* The key characteristic is that these elements cannot be understood in isolation or randomly reordered without losing critical information about their relationships and meaning.

**Modelling Sequence Data problems :**
*	In traditional neural networks, all the inputs and outputs are independent of each other, but in case of sequential data like when it is required to predict the next word of a sentence, the previous words are required and hence there is a need to remember the previous words.
*	Thus **RNN came into existence**, which solved this issue with the help of a **Hidden state**. The main and most important feature of RNN is Hidden state, which remembers some information about a sequence.

What is our aim : Instead of considering timesteps in isolation , we want the output at a later time stamp Yt , to depend on input in the oberveation we saw prior.

**How to model that relation ?**

* One idea is to introduce a new variable representing the **state of neural network --> Ht**

* This state Ht is learnt and computed by neurons in the layer and that is passed to next timestamp

image.png

* here the **state ht** is updated sequentially and iteratively .

* So the state ht is essentially the "memory" or "internal representation" of the RNN that gets updated at each time step

* and also we can understand that now the **output Yt at a time stamp** depends on **input at that time stamp Xt** and hidden state of neural network at previous time stamp **h(t-1)**

image.png

* This idea of passing the state forwards to a time is the basis of what we call the **recuurance cell, or neurons with recurrance.**.
