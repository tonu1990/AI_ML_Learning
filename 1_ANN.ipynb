## **Introduction : What is Deep Learning?**

* Deep learning is a branch of machine learning which is completely based on
artificial neural networks, as neural network is going to mimic the human brain so deep learning is also a kind of mimic of human brain.
* In deep learning, we don’t need to explicitly program everything.
* The concept of deep learning is not new. It has been around for a couple of years now. It’s on hype nowadays because earlier we did not have that much processing power and a lot of data. As in the last 25 years, the processing power increases exponentially, deep learning and machine learning came in the picture.
* Generally speaking, deep learning is a machine learning method that takes in an input X, and uses it to predict an output of Y.
  * As an example, given the stock prices of the past week as input, my deep learning algorithm will try to predict the stock price of the next day.
  * 	As another example, let’s say that inputs are images of dogs and cats, and outputs are labels for those images (i.e. is the input picture a dog or a cat). If an input has a label of a dog, but the deep learning algorithm predicts a cat, then my deep learning algorithm will learn that the features of my given image (e.g. sharp teeth, facial features) are going to be associated with a dog.
*	Given a large dataset of input and output pairs, a deep learning algorithm will try to minimize the difference between its prediction and expected output. By doing this, it tries to learn the association/pattern between given inputs and outputs — this in turn allows a deep learning model to generalize to inputs that it hasn’t seen before.



**Deep Learning models can be used for a variety of complex tasks ;**

* Artificial Neural Networks (ANNs) → Regression and Classification
  * Used for structured data, tabular data, and general prediction tasks.Works well when feature engineering is important.

* Convolutional Neural Networks (CNNs) → Computer Vision
  * Used for image classification, object detection, segmentation (e.g., YOLO, ResNet, U-Net).Also applied in medical imaging and remote sensing.

* Recurrent Neural Networks (RNNs) → Time Series Analysis
  * Used for sequential data (e.g., stock price prediction, speech recognition).Variants like LSTMs and GRUs handle long-term dependencies better.

* Self-Organizing Maps (SOMs) → Feature Extraction
  * Unsupervised learning for clustering and dimensionality reduction.Used in anomaly detection, data visualization, and market segmentation.

* Deep Boltzmann Machines (DBMs) → Recommendation Systems
  * DBMs are generative models, but they are not commonly used in modern recommendation systems.Instead, Deep Neural Networks (DNNs), Autoencoders, Graph Neural Networks (GNNs), and Transformers are more widely used.

* Autoencoders → Feature Learning, Dimensionality Reduction, Anomaly Detection.
  * Often used in recommendation systems, but more for anomaly detection, denoising, and feature extraction.Variants like Variational Autoencoders (VAEs) and Denoising Autoencoders are powerful alternatives.

* Transformers → NLP, Vision, and Multi-Modal Learning
  * Revolutionized NLP with models like BERT, GPT, and T5.
  * Vision Transformers (ViTs) compete with CNNs in computer vision tasks.
  * Applied in multi-modal tasks like combining text and image understanding.

* Graph Neural Networks (GNNs) → Social Networks, Fraud Detection,Bioinformatics
  * Powerful for recommendation systems, fraud detection, and molecular property prediction.

* Generative Adversarial Networks (GANs) → Image Synthesis, Data Augmentation
  * Used in deepfake generation, medical imaging, and artistic style transfer.

* Diffusion Models → Generative AI
  * Emerging as powerful alternatives to GANs for image and video generation (e.g., Stable Diffusion).

 **Single neuron gets multiple input signals and sends single output**

* Each input will be a value representing the feature (like age, income etc..)
* We have weights associated with each input. Initially they will be assigned at random.
* Then inputs are multiplied with **weights** , and their added sume is then passed to **activation function** .

**Concept of weights:**

* Weights are the main feature of neural network learning.
* These weights are often initially created by random values.
* NN learn by adjusting the weights.
* They decide which input signal is more important and which is less.
* They decide which signal get passed along and which are not.
* They get adjusted during the process of learning (gradient descent and back propagation comes into picture there).
* As learning occurs, these weights are changed (the model is learning), until the model reaches convergence (it does not always reach convergence).

**Activation function :**

* They basically decide whether a neuron should be activated or not.
* This function is applied on the weighted sum of inputs, and the output of function is passed to next level.
* There are different activation functions depending on the location of the perceptron (inner hidden layers or output layer), and the problem context.

**Single Perceptron to multi Layer perceptron**

* The concept of a single perceptron, which involves inputs, weights, a weighted sum, and an activation function, serves as the foundational building block for more complex neural network architectures.
* While a single perceptron can effectively model linear relationships, it is limited in its ability to capture non-linear patterns.
* To overcome this limitation, the architecture can be extended to a **multi-layer perceptron** (MLP), also known as a **feedforward neural network**.
* In an MLP, multiple layers of perceptrons are stacked, with each layer's output serving as the input to the next. This layered structure allows the **network to learn hierarchical representations of data, where each successive layer captures increasingly abstract features**.
* The **term "deep learning**" originates from the use of many such layers, or "depth," in the network, enabling the model to learn complex patterns and representations.
* A fully connected network, or dense network, refers to a type of neural network where each neuron in one layer is connected to every neuron in the subsequent layer. This dense connectivity ensures that the network can learn a wide variety of features and interactions between inputs.
* The combination of deep architectures and fully connected layers has been instrumental in the success of deep learning, allowing models to achieve state-of-the-art performance in tasks such as image recognition, natural language processing, and more.

**What is Gradient ?**





* A derivative of 0.0 indicates no change in the target function, referred to as a stationary point.

* A function may have one or more stationary points and a local or global minimum (bottom of a valley) or maximum (peak of a mountain) of the function are examples of stationary points.

**Gradient in Neural Networks**
* In the context of neural networks, gradients are used in the training process to **minimize the error of the model**. This is typically done using a method called gradient descent.

* Here's a simple explanation of how it works:
    * Initialization: Start with random weights for the connections in the neural network.
    * Forward Pass: Input data is passed through the network to get predictions.
    * Compute Loss: Calculate the error (or loss) by comparing the predictions to the actual target values.
    * Backward Pass (Backpropagation): **Compute the gradient of the loss with respect to each weight in the network**. This tells us how much the loss would change if we changed the weights slightly.
    * Update Weights: Adjust the weights in the opposite direction of the gradient to reduce the loss. This is done using a learning rate, which controls how big the update steps are.
    * Repeat: This process is repeated for many iterations until the model's predictions are sufficiently accurate.

**What is vanishing gradient problem ?**

* The vanishing gradient problem is an issue that can occur during the training of deep neural networks, particularly those with many layers.
* It refers to the situation where the gradients of the loss function with respect to the weights become very small as they are propagated back through the network.
* the gradients tend to become smaller as they are propagated backward through the layers.
* This can lead to extremely slow learning or even prevent the network from learning effectively.
*

**Activation functions used in HIDDEN LAYERS :**

1. Sigmoid function
    

Characteristics:
* Outputs values between 0 and 1.
* Can cause the **vanishing gradient problem, especially in deep networks**, because the gradients become very small for large positive or negative inputs.
* slow convergence
* Use Case: While historically used in hidden layers, it's less common now due to the vanishing gradient issue. It might still be used in simpler or shallower networks.

2. Tan H


Characteristics:
* Outputs values between -1 and 1, which makes it zero-centered.
* Like Sigmoid, it can also suffer from the vanishing gradient problem, but to a lesser extent.
* Faster convergence than Sigmoid
* Use Case: Often preferred over Sigmoid for hidden layers because it is zero-centered, which can help with convergence. It is suitable for networks where zero-centered outputs are beneficial.
* The hyperbolic tangent function is a popular choice for the hidden layer activation in RNNs. It outputs values between -1 and 1, which helps in maintaining a stable range of activations and can capture both positive and negative dependencies.

3. ReLU


Characteristics:
* Outputs values between 0 and infinity.
* Helps mitigate the vanishing gradient problem, allowing for faster training and deeper networks.
* Can suffer from the "dying ReLU" problem, where neurons become inactive.
* Use Case: Widely used in hidden layers of deep neural networks due to its simplicity and effectiveness. It is a good default choice for most applications.
